```python
import geopandas as gpd
import pandas as pd
import numpy as np
import folium
```

```python
df = pd.read_csv('visualize-data.csv')
df.head()
# df.dtypes
```

```python
df = df [['fire_size', 'latitude', 'longitude', 'state', 'stat_cause_descr']]
x = len(df)
```

```python
df.head(15)
```

```python
df = df.loc[df['fire_size'] > 1000]
```

```python
m = folium.Map(location = [37.256500, -122.027153],
               zoom_start = 6.5)

small_wildfires = folium.FeatureGroup(name = '< 10000 Acres')
medium_wildfires = folium.FeatureGroup(name = '10,000 - 50,000 Acres')
large_wildfires = folium.FeatureGroup(name = '50,000 - 100,000 Acres')
xl_wildfires = folium.FeatureGroup(name = '> 100,000 Acres')
lightning = folium.FeatureGroup(name = 'Lightning')
arson = folium.FeatureGroup(name = 'Arson')
miscellaneous = folium.FeatureGroup(name = 'Miscellaneous')



for i, v in df.iterrows():
    fire_size = float(v['fire_size'])
    state = str(v['state'])
    cause = str(v['stat_cause_descr'])
    
    popup = """
    Size (Acres) : <b>%s</b><br>
    State : <b>%s</b><br>
    Cause : <b>%s</b><br>
    """ % (v['fire_size'], v['state'], v['stat_cause_descr'])
    
    
    if fire_size < 10000:
        folium.CircleMarker(location = [v['latitude'], v['longitude']],
                           radius = np.log(fire_size) * 0.8,
                           weight = 0,
                           tooltip = popup,
                           color = '#ffeda0',
                           fill_color = '#ffeda0',
                           fill_opacity = 0.7,
                           fill = True).add_to(small_wildfires)
        
    if fire_size in range(10000, 50000):
        folium.CircleMarker(location = [v['latitude'], v['longitude']],
                           radius = np.log(fire_size),
                           weight = 0,
                           tooltip = popup,
                           color = '#feb24c',
                           fill_color = '#feb24c',
                           fill_opacity = 0.7,
                           fill = True).add_to(medium_wildfires)
    
    if fire_size in range(50000, 100000):
        folium.CircleMarker(location = [v['latitude'], v['longitude']],
                           radius = np.log(fire_size) * 1.5,
                           weight = 0,
                           tooltip = popup,
                           color = '#fc4e2a',
                           fill_color = '#fc4e2a',
                           fill_opacity = 0.7,
                           fill = True).add_to(large_wildfires)

    if fire_size > 100000:
        folium.CircleMarker(location = [v['latitude'], v['longitude']],
                           radius = np.log(fire_size) * 2,
                           weight = 0,
                           tooltip = popup,
                           color = '#b10026',
                           fill_color = '#b10026',
                           fill_opacity = 0.7,
                           fill = True).add_to(xl_wildfires)

small_wildfires.add_to(m)
medium_wildfires.add_to(m)
large_wildfires.add_to(m)
xl_wildfires.add_to(m)
folium.LayerControl(collapsed = False).add_to(m)

m.save(r'templates/visualize-map.html')
m
```

```python
m2 = folium.Map(location = [37.256500, -122.027153],
               zoom_start = 6.5)

lightning = folium.FeatureGroup(name = 'Lightning')
arson = folium.FeatureGroup(name = 'Arson')
miscellaneous = folium.FeatureGroup(name = 'Miscellaneous')
debris = folium.FeatureGroup(name = 'Debris Burning')
equipment = folium.FeatureGroup(name = 'Equipment Use')
campfire = folium.FeatureGroup(name = 'Campfire')



for i, v in df.iterrows():
    fire_size = float(v['fire_size'])
    state = str(v['state'])
    cause = str(v['stat_cause_descr'])
    
    popup = """
    Size (Acres) : <b>%s</b><br>
    State : <b>%s</b><br>
    Cause : <b>%s</b><br>
    """ % (v['fire_size'], v['state'], v['stat_cause_descr'])
    
    
    if cause == "Lightning":
        folium.CircleMarker(location = [v['latitude'], v['longitude']],
                           radius = np.log(fire_size) * 0.8,
                           weight = 0,
                           tooltip = popup,
                           color = '#0000FF',
                           fill_color = '#0000FF',
                           fill_opacity = 0.7,
                           fill = True).add_to(lightning)
        
    if cause == "Miscellaneous":
        folium.CircleMarker(location = [v['latitude'], v['longitude']],
                           radius = np.log(fire_size),
                           weight = 0,
                           tooltip = popup,
                           color = '#feb24c',
                           fill_color = '#feb24c',
                           fill_opacity = 0.7,
                           fill = True).add_to(miscellaneous)
    
    if cause == "Arson":
        folium.CircleMarker(location = [v['latitude'], v['longitude']],
                           radius = np.log(fire_size) * 1.5,
                           weight = 0,
                           tooltip = popup,
                           color = '#b10026',
                           fill_color = '#b10026',
                           fill_opacity = 0.7,
                           fill = True).add_to(arson)

    if cause == "Debris Burning":
         folium.CircleMarker(location = [v['latitude'], v['longitude']],
                           radius = np.log(fire_size) * 1.5,
                           weight = 0,
                           tooltip = popup,
                           color = '#e534eb',
                           fill_color = '#e534eb',
                           fill_opacity = 0.7,
                           fill = True).add_to(debris)

    if cause == "Equipment Use":
         folium.CircleMarker(location = [v['latitude'], v['longitude']],
                           radius = np.log(fire_size) * 1.5,
                           weight = 0,
                           tooltip = popup,
                           color = '#747eed',
                           fill_color = '#747eed',
                           fill_opacity = 0.7,
                           fill = True).add_to(equipment)

    if cause == "Campfire":
         folium.CircleMarker(location = [v['latitude'], v['longitude']],
                           radius = np.log(fire_size) * 1.5,
                           weight = 0,
                           tooltip = popup,
                           color = '#ffce63',
                           fill_color = '#ffce63',
                           fill_opacity = 0.7,
                           fill = True).add_to(campfire)
lightning.add_to(m2)
arson.add_to(m2)
miscellaneous.add_to(m2)
debris.add_to(m2)
equipment.add_to(m2)
campfire.add_to(m2)
folium.LayerControl(collapsed = False).add_to(m2)
m2.save(r'templates/visualize-map2.html')
m2
```
